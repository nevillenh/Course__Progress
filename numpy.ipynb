```python
# Load a dataset
import pandas as pd
df = pd.read_csv("teslastockdata.csv")
print(df.head())
```

```python
# Extract the "Close" column as a NumPy array
close_prices = df["Close"].to_numpy()

print(close_prices)
```

```python
# Some basic statistics
import numpy as np
mean_price = np.mean(close_prices)
print("Average Close Price:", mean_price)
```

```python
# high_price (and when it happened)
max_price = np.max(close_prices)
max_index = np.argmax(close_prices)

print("Highest Price:", max_price)
print("Day Index:", max_index)
print("Date of Highest Price:", df["Date"][max_index])
```

```python
# Daily returns
returns = np.diff(close_prices) / close_prices[:-1]

print("Daily Returns:", returns[:5])  # first 5 returns
```

```python
# Measure of volatility
volatility = np.std(returns)

print("Volatility:", volatility)
```

```python
# PLOT THE STOCK TREND
import matplotlib.pyplot as plt

plt.plot(close_prices)
plt.title("Stock Price Over Time")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()
```
