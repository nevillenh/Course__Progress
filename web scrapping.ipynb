```python
# WEB SCRAPING

import requests
from bs4 import BeautifulSoup, Comment
# URL of the website to scrape
url = 'https://www.aiu.ac.ke/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# Extracting all the links from the webpage
links = soup.find_all('a')
for link in links:    print(link.get('href'))
```

```python
# Extracting all the paragraphs from the webpage
paragraphs = soup.find_all('p')
for paragraph in paragraphs:    print(paragraph.text)
```

```python
# Extracting all the headings from the webpage
headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
for heading in headings:    print(heading.text)
```

```python
# Extracting all the list items from the webpage
list_items = soup.find_all('li')
for item in list_items:    print(item.text) 
```

```python
# Extracting all the tables from the webpage
tables = soup.find_all('table')
for table in tables:    print(table.text)
```

```python
# Extracting all the forms from the webpage
forms = soup.find_all('form')
for form in forms:    print(form.text) 
```

```python
# Extracting all the divs from the webpage
divs = soup.find_all('div')
for div in divs:    print(div.text)
```

```python
# Extracting all the spans from the webpage
spans = soup.find_all('span')
for span in spans:    print(span.text)
```

```python
# Extracting all the scripts from the webpage
scripts = soup.find_all('script')
for script in scripts:    print(script.text) 
```

```python
# saving the webpage content to a file
with open('webpage_content.txt', 'w', encoding='utf-8') as file:
    file.write(soup.prettify()) 

```

```python
# find classes
classes = soup.find_all(class_='class-name')
for class_ in classes:    print(class_.text)
```
